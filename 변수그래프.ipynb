```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import platform
pd.options.display.float_format = '{:.5f}'.format

if platform.system() == 'Windows':
    matplotlib.rc('font', family='Malgun Gothic')
elif platform.system() == 'Darwin': # Mac
    matplotlib.rc('font', family='AppleGothic')
else: #linux
    matplotlib.rc('font', family='NanumGothic')



co = pd.read_csv('../data_chaeyeon/최종파일들/모아두기/CO오염도_최종.csv')
car_nor = pd.read_csv('../data_chaeyeon/최종파일들/모아두기/일반자동차_최종.csv')
car_elec = pd.read_csv('../data_chaeyeon/최종파일들/모아두기/전기자동차_최종.csv')



co_data = co.iloc[0:, 1:]
car_nor_data = car_nor.iloc[0:, 1:]
car_elec_data = car_elec.iloc[0:, 1:]


coDivcar_elec = co_data.div(car_elec_data, fill_value=0)

do = pd.DataFrame({
    '지역' : ['서울', '경기', '인천', '대구', '제주']
})
coDivcar_elec_final = pd.concat([do, coDivcar_elec], axis=1)
coDivcar_elec_final_a = ['2018년1월','2018년2월','2018년3월','2018년4월','2018년5월','2018년6월','2018년7월','2018년8월','2018년9월','2018년10월','2018년11월','2018년12월',
                        '2019년1월','2019년2월','2019년3월','2019년4월','2019년5월','2019년6월','2019년7월','2019년8월','2019년9월','2019년10월','2019년11월','2019년12월',
                        '2020년1월','2020년2월','2020년3월','2020년4월','2020년5월','2020년6월','2020년7월','2020년8월','2020년9월','2020년10월','2020년11월', '2020년12월']

coDivcar_elec_final_seoul = coDivcar_elec_final[coDivcar_elec_final['지역']=='서울']
coDivcar_elec_final_seoul_a = coDivcar_elec_final_seoul.iloc[0: ,1: ]
coDivcar_elec_final_seoul_a = coDivcar_elec_final_seoul_a.values.reshape(36,1)

coDivcar_elec_final_gg = coDivcar_elec_final[coDivcar_elec_final['지역']=='경기']
coDivcar_elec_final_gg_a = coDivcar_elec_final_gg.iloc[0: ,1: ]
coDivcar_elec_final_gg_a = coDivcar_elec_final_gg_a.values.reshape(36,1)

coDivcar_elec_final_incheon = coDivcar_elec_final[coDivcar_elec_final['지역']=='인천']
coDivcar_elec_final_incheon_a = coDivcar_elec_final_incheon.iloc[0: ,1: ]
coDivcar_elec_final_incheon_a = coDivcar_elec_final_incheon_a.values.reshape(36,1)

coDivcar_elec_final_dg = coDivcar_elec_final[coDivcar_elec_final['지역']=='대구']
coDivcar_elec_final_dg_a = coDivcar_elec_final_dg.iloc[0: ,1: ]
coDivcar_elec_final_dg_a = coDivcar_elec_final_dg_a.values.reshape(36,1)

coDivcar_elec_final_jj = coDivcar_elec_final[coDivcar_elec_final['지역']=='제주']
coDivcar_elec_final_jj_a = coDivcar_elec_final_jj.iloc[0: ,1: ]
coDivcar_elec_final_jj_a = coDivcar_elec_final_jj_a.values.reshape(36,1)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot()

ax.plot(coDivcar_elec_final_a,coDivcar_elec_final_seoul_a, marker='o',label='서울')
ax.plot(coDivcar_elec_final_a,coDivcar_elec_final_gg_a, marker='o',label='경기')
ax.plot(coDivcar_elec_final_a,coDivcar_elec_final_incheon_a, marker='o',label='인천')
ax.plot(coDivcar_elec_final_a,coDivcar_elec_final_dg_a, marker='o',label='대구')
ax.plot(coDivcar_elec_final_a,coDivcar_elec_final_jj_a, marker='o',label='제주')
plt.ylim([0, 0.008])

ax.legend()
plt.xticks(rotation=45)
plt.title('',fontsize=20)
plt.grid(True, alpha=0.2, linestyle='--')
plt.show()
```
